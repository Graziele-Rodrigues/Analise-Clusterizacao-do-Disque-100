```python
df2 = df.copy()

v = df2["violacao"].astype(str).str.split(">", expand=True)
df2["violacao_n1"] = v[0]
df2["violacao_n2"] = v[1]
df2["violacao_n3"] = v[2]


def moda(series):
    series = series.dropna()
    if len(series) == 0:
        return np.nan
    return series.value_counts().idxmax()

def agrega_por_hash(df, id_col="hash", num_cols=None, bool_like_cols=None):
    if num_cols is None: num_cols = []
    if bool_like_cols is None: bool_like_cols = []

    agg = {}
    for c in df.columns:
        if c == id_col:
            continue
        if c in num_cols:
            agg[c] = "max"
        elif c in bool_like_cols:
            agg[c] = lambda s: "SIM" if (s.astype(str).str.upper() == "SIM").any() else "NÃO"
        else:
            agg[c] = moda

    return df.groupby(id_col, as_index=False).agg(agg)


num_cols = ["sl_quantidade_vitimas"]

bool_like_cols = [
    "Denúncia_emergencial",
    "Deficiência_da_vítima", "Doença_rara_da_vítima",
    "Deficiência_do_suspeito", "Doença_rara_do_suspeito",
]

df_hash = agrega_por_hash(df2, num_cols=num_cols, bool_like_cols=bool_like_cols)
df_hash.head()
print(',\n'.join(df_hash.columns))
```

```python
# transformacao numerica
from sklearn.preprocessing import StandardScaler

def prepara_matrix(df_in, num_cols=None):
    if num_cols is None: num_cols = []
    df_in = df_in.copy()

    # separa numéricas
    df_num = df_in[num_cols].copy() if len(num_cols) else pd.DataFrame(index=df_in.index)
    df_cat = df_in.drop(columns=num_cols, errors="ignore")

    # one-hot
    X_cat = pd.get_dummies(df_cat, dummy_na=True)

    # normaliza numéricas
    if len(num_cols):
        scaler = StandardScaler()
        df_num[num_cols] = scaler.fit_transform(df_num[num_cols])

    X = pd.concat([X_cat, df_num], axis=1).fillna(0)
    return X

X_S = prepara_matrix(df_S, num_cols=["sl_quantidade_vitimas"])
X_P = prepara_matrix(df_P, num_cols=[])
```

## Clusterização para perfil vitima/suspeito



```python
# Copiando os dados originais (one-hot binário)
X = X_P.copy()
```

```python
nComponents = 15
pca_exploratorio = PCA(n_components=nComponents, random_state=42)
pca_exploratorio.fit(X)

explainedVariance = pca_exploratorio.explained_variance_ratio_
cumulativeExplainedVariance = np.cumsum(explainedVariance)

resultPCA = pd.DataFrame({
    "Componente": np.arange(1, nComponents + 1),
    "Variacao_Explicada": explainedVariance,
    "Variacao_Acumulada": cumulativeExplainedVariance
})

print(resultPCA.to_string(index=False))
```

```python
# Copiando variáveis necessárias
explainedVariance412 = explainedVariance

# Contribuição incremental de cada componente adicional e a identificar o ponto de cotovelo
plt.figure(figsize=(8, 4))
plt.plot(np.cumsum(explainedVariance412))
plt.xlabel('Número de Componentes')
plt.ylabel('Variação Cumulativa Explicada')
plt.title('Variação Explicada Acumulada pelo PCA')
plt.axhline(y=0.90, color='r', linestyle='--')
plt.text(0.5, 0.85, '90% de variação', color = 'red')
plt.grid(True)
plt.show()
```

```python
# Copiando variáveis necessárias
explainedVariance413 = explainedVariance
nComponents423 = nComponents

# Criando um gráfico de barras para a variação explicada por cada componente
plt.figure(figsize=(10, 6))
plt.bar(range(1, nComponents423+1), explainedVariance413, alpha=0.7, align='center', label='Variação individual explicada')
plt.step(range(1, nComponents423+1), np.cumsum(explainedVariance413), where='mid', label='Variação acumulada explicada')
plt.ylabel('Razão da Variação Explicada')
plt.xlabel('Componentes Principais')
plt.legend(loc='best')
plt.xticks(range(1, nComponents423+1))
plt.title('Variação explicada por cada componente')
plt.show()
```

```python
pca_final = PCA(n_components=0.90, random_state=42)  # mantém 90% da variância
X_PCA = pca_final.fit_transform(X)

print(f"Número final de componentes: {pca_final.n_components_}")
```

**Número Ideal de Cluster**

```python
# Copiando variáveis necessárias
df_perfil = X_PCA.copy()

# Determina o número ideal de clusters usando o método do cotovelo
wcss = []
maxClusters = 15
for i in range(1, maxClusters+1):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=42)

    kmeans.fit(df_perfil)
    wcss.append(kmeans.inertia_)

# Criando um DataFrame com a quantidade de clusters e WCSS
resultWCSS = pd.DataFrame({'Number of Clusters': range(1, maxClusters + 1), 'WCSS': wcss})

# Exibindo o DataFrame
print(resultWCSS.to_string(index=False))
```

### Agrupamento das colunas ONE-HOT

```python
import numpy as np

def cols_like(df, prefix, contains_any=None, exclude_any=None):
    cols = [c for c in df.columns if c.startswith(prefix)]
    if contains_any:
        cols = [c for c in cols if any(x in c for x in contains_any)]
    if exclude_any:
        cols = [c for c in cols if not any(x in c for x in exclude_any)]
    return cols

def flag_from_any(df, cols):
    if not cols:
        return 0
    return df[cols].fillna(0).astype(int).any(axis=1).astype(int)

def safe_get01(df, col):
    if col not in df.columns:
        return 0
    return df[col].fillna(0).astype(int).clip(0,1)


def build_flags(df, cluster_col="Cluster"):
    out = df.copy()

    # -----------------------
    # Sexo vítima
    # -----------------------
    out["vitima_mulher"]    = safe_get01(out, "sexo_da_vitima_FEMININO")
    out["vitima_homem"]     = safe_get01(out, "sexo_da_vitima_MASCULINO")

    outros_sexo = [
        "sexo_da_vitima_NÃO INFORMADO",
        "sexo_da_vitima_NÃO SE APLICA - VÍTIMA COMUNIDADE/FAMÍLIA",
        "sexo_da_vitima_nan"
    ]
    out["vitima_sexo_outros"] = flag_from_any(out, [c for c in outros_sexo if c in out.columns])

    # -----------------------
    # Faixa etária (one-hot)
    # -----------------------
    crianca_tokens = [
        "RECÉM", "MENOS",
        "01", "02", "03", "04", "05", "06", "07", "08", "09",
        "10", "11", "12", "13", "14", "15", "16", "17"
    ]
    idoso_tokens = ["60", "65", "70", "75", "80", "85", "90+", "IDOSO"]

    crianca_cols = cols_like(out, "faixa_etaria_da_vitima_", contains_any=crianca_tokens)
    idoso_cols   = cols_like(out, "faixa_etaria_da_vitima_", contains_any=idoso_tokens)

    out["vitima_crianca"] = flag_from_any(out, crianca_cols)
    out["vitima_idoso"]   = flag_from_any(out, idoso_cols)

    # -----------------------
    # Deficiência vítima
    # -----------------------
    # Preferir SIM explícito (se existir)
    if "deficiencia_da_vitima_SIM" in out.columns:
        out["vitima_deficiencia"] = safe_get01(out, "deficiencia_da_vitima_SIM")
    elif "deficiencia_da_vitima_NÃO" in out.columns:
        # só use se for binário (SIM/NÃO). caso contrário, pode distorcer.
        out["vitima_deficiencia"] = (1 - out["deficiencia_da_vitima_NÃO"].fillna(0)).clip(0,1).astype(int)
    elif "vitima_com_deficiencia" in out.columns:
        out["vitima_deficiencia"] = safe_get01(out, "vitima_com_deficiencia")
    else:
        out["vitima_deficiencia"] = 0

    # -----------------------
    # Suspeito
    # -----------------------
    out["suspeito_homem"]  = safe_get01(out, "sexo_do_suspeito_MASCULINO")
    out["suspeito_mulher"] = safe_get01(out, "sexo_do_suspeito_FEMININO")
    out["suspeito_nao_informado"] = (
    safe_get01(out, "genero_do_suspeito_NÃO INFORMADO") |
    safe_get01(out, "genero_do_suspeito_NÃO SE APLICA - VÍTIMA COMUNIDADE/FAMÍLIA") |
    safe_get01(out, "genero_do_suspeito_nan")
    ).astype(int)


    # -----------------------
    # Relação
    # -----------------------
    rel_familiar = ["PAI", "MÃE", "FILHO", "IRMÃO", "AVÔ", "NETO", "PADRASTO", "MADRASTA"]
    rel_conjugal = ["COMPANHEIRO", "NAMORADO", "ESPOSA", "EX-"]
    rel_inst     = ["PROFESSOR", "DIRETOR", "GESTOR", "INSTITUIÇÃO", "FUNCIONÁRIO"]

    fam_cols = cols_like(out, "relacao_vitima_suspeito_", contains_any=rel_familiar)
    conj_cols = cols_like(out, "relacao_vitima_suspeito_", contains_any=rel_conjugal)
    inst_cols = cols_like(out, "relacao_vitima_suspeito_", contains_any=rel_inst)

    out["rel_familiar"] = flag_from_any(out, fam_cols)
    out["rel_conjugal"] = flag_from_any(out, conj_cols)
    out["rel_inst"]     = flag_from_any(out, inst_cols)

    # se quiser "desconhecido" explícito
    out["rel_desconhecido"] = safe_get01(out, "relacao_vitima_suspeito_DESCONHECIDO(A)")

    # -----------------------
    # Orientação
    # -----------------------
    orient_cols = cols_like(out, "orientacao_sexual_da_vitima_")
    lgbtq_tokens = ["HOMOSSEXUAL", "BISSEXUAL", "TRANS", "OUTRO"]
    lgbtq_cols = [c for c in orient_cols if any(x in c for x in lgbtq_tokens)]
    out["vitima_lgbtq"] = flag_from_any(out, lgbtq_cols)

    nao_info_cols = [
        "orientacao_sexual_da_vitima_NÃO INFORMADO",
        "orientacao_sexual_da_vitima_nan"
    ]
    out["vitima_orientacao_nao_info"] = flag_from_any(out, [c for c in nao_info_cols if c in out.columns])

    # -----------------------
    # Raça/cor
    # -----------------------
    out["vitima_branca"] = safe_get01(out, "raca_cor_da_vitima_BRANCA")
    out["vitima_preta_parda"] = (
        safe_get01(out, "raca_cor_da_vitima_PRETA") |
        safe_get01(out, "raca_cor_da_vitima_PARDA")
    ).astype(int)
    out["vitima_indigena"] = safe_get01(out, "raca_cor_da_vitima_INDÍGENA")

    raca_nao_info = ["raca_cor_da_vitima_NÃO INFORMADO", "raca_cor_da_vitima_nan"]
    out["vitima_raca_nao_info"] = flag_from_any(out, [c for c in raca_nao_info if c in out.columns])

    # garante cluster existe
    if cluster_col not in out.columns:
        raise ValueError(f"Coluna {cluster_col} não existe.")
    return out
```

```python
def cluster_profile_means(df_flags, cluster_col="Cluster", flag_prefixes=None):
    # por padrão pega as flags criadas
    if flag_prefixes is None:
        flag_prefixes = ("vitima_", "suspeito_", "rel_")

    flag_cols = [c for c in df_flags.columns if c.startswith(flag_prefixes)]
    prof = df_flags.groupby(cluster_col)[flag_cols].mean()

    # renomeia para pct_
    prof = prof.rename(columns={c: f"pct_{c}" for c in prof.columns})

    # deixa índice ordenado mas sem assumir int
    prof = prof.sort_index()
    return prof

df_flags = build_flags(dfClusteredB, cluster_col="Cluster")
dfClusterMeans = cluster_profile_means(df_flags, cluster_col="Cluster")
```

```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np

def plot_bubble_chart_sim_only(df, y_attrs, x_attrs, cluster_col='Cluster'):
    # validações
    y_attrs = [c for c in y_attrs if c in df.columns]
    x_attrs = [c for c in x_attrs if c in df.columns]
    if not y_attrs or not x_attrs:
        raise ValueError("y_attrs ou x_attrs vazios.")
    if cluster_col not in df.columns:
        raise ValueError(f"Coluna {cluster_col} não existe.")

    clusters = sorted(df[cluster_col].dropna().unique())
    colors = ['lightcoral', 'skyblue', 'gold', 'limegreen', 'mediumpurple']

    fig = go.Figure()

    y_spacing = 0.8
    bubble_spacing = 0.6

    # posição base de cada atributo
    y_positions = {attr: i * (len(clusters) * bubble_spacing + y_spacing)
                   for i, attr in enumerate(y_attrs)}

    for c in clusters:
        dfc = df[df[cluster_col] == c]

        rows = []
        for y in y_attrs:
            for x in x_attrs:
                count_pos = int(((dfc[y] == 1) & (dfc[x] == 1)).sum())
                y_pos = y_positions[y] + clusters.index(c) * bubble_spacing
                rows.append((x, y_pos, y, count_pos))

        d = pd.DataFrame(rows, columns=["x", "y", "y_label", "count"])

        # evita bolhas invisíveis
        sizes = np.sqrt(d["count"].values)
        sizes = np.maximum(sizes, 0.3)

        sizeref = sizes.max() / (45.0**2) if sizes.max() > 0 else 1

        fig.add_trace(go.Scatter(
            x=d["x"],
            y=d["y"],
            mode="markers+text",
            marker=dict(
                size=sizes,
                sizemode="area",
                sizeref=sizeref,
                color=colors[c % len(colors)],
                opacity=0.6,
                line=dict(width=0.5, color="rgba(0,0,0,0.3)")
            ),
            text=d["count"].where(d["count"] > 0, ""),
            textposition="middle center",
            textfont=dict(color="black"),
            hovertemplate=(
                "<b>Relação:</b> %{x}<br>"
                "<b>Atributo da vítima:</b> %{customdata[0]}<br>"
                "<b>Quantidade:</b> %{customdata[1]}<br>"
                "<b>Cluster:</b> " + str(c) +
                "<extra></extra>"
            ),
            customdata=np.stack([d["y_label"], d["count"]], axis=1),
            name=f"Cluster {c}"
        ))

    # ticks do eixo Y
    y_tickvals = [
        y_positions[a] + ((len(clusters) - 1) * bubble_spacing) / 2
        for a in y_attrs
    ]

    fig.update_layout(
        title='Gráfico de Bolhas por Cluster',
        xaxis=dict(
            tickmode="array",
            tickvals=x_attrs,
            ticktext=x_attrs,
            tickangle=30,
            showgrid=False,
            zeroline=False
        ),
        yaxis=dict(
            tickmode="array",
            tickvals=y_tickvals,
            ticktext=y_attrs,
            showgrid=False,
            zeroline=False
        ),
        height=260 + 90 * len(y_attrs),
        width=300 + 200 * len(x_attrs),
        showlegend=True
    )

    fig.show()


df_for_plot = df_flags.copy()

atributos_Y = [
    "vitima_mulher",
    "vitima_crianca",
    "vitima_idoso",
    "vitima_preta_parda",
    "vitima_lgbtq",
    "vitima_deficiencia",
]

atributos_X = [
    "rel_familiar",
    "rel_conjugal",
    "rel_inst",
    "rel_desconhecido",
]

plot_bubble_chart_sim_only(df_for_plot, atributos_Y, atributos_X)
```

```python
import re
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def onehot_to_label(df, prefix):
    cols = [c for c in df.columns if c.startswith(prefix)]
    if not cols:
        raise ValueError(f"Nenhuma coluna encontrada com prefixo {prefix}")

    sub = df[cols].fillna(0)
    idx = sub.values.argmax(axis=1)
    maxv = sub.values.max(axis=1)

    labels = pd.Series([cols[i].replace(prefix, "") for i in idx], index=df.index)
    labels[maxv == 0] = "SEM INFO"
    return labels

def age_group(label: str) -> str:
    if label is None:
        return "SEM INFO"
    s = str(label).upper().strip()

    # sem info
    if s in ["SEM INFO", "NÃO INFORMADO", "NÃO INFORMADA", "NAN"] or "NÃO INFORM" in s:
        return "SEM INFO"

    # recém / menos de 1 ano
    if "RECÉM" in s or "MENOS" in s:
        return "CRIANÇA (0–11)"

    # pega o primeiro número que aparecer
    m = re.search(r"(\d+)", s)
    if not m:
        # casos tipo "CRIANÇA/ADOLESCENTE IDADE NÃO INFORMADA"
        if "CRIANÇA" in s:
            return "CRIANÇA (0–11)"
        if "IDOSO" in s:
            return "IDOSO (60+)"
        return "SEM INFO"

    age = int(m.group(1))

    # 90+
    if "90" in s and "+" in s:
        age = 90

    if age <= 11:
        return "CRIANÇA (0–11)"
    if 12 <= age <= 17:
        return "ADOLESCENTE (12–17)"
    if 18 <= age <= 29:
        return "JOVEM (18–29)"
    if 30 <= age <= 59:
        return "ADULTO (30–59)"
    return "IDOSO (60+)"

def heatmap_agegroup_vitima_vs_suspeito(df, cluster=None, cluster_col="Cluster", normalize="row"):
    d = df.copy()
    if cluster is not None:
        d = d[d[cluster_col] == cluster].copy()

    vit_raw = onehot_to_label(d, "faixa_etaria_da_vitima_")
    sus_raw = onehot_to_label(d, "faixa_etaria_do_suspeito_")

    vit = vit_raw.map(age_group)
    sus = sus_raw.map(age_group)

    tab = pd.crosstab(sus, vit)

    # ordena grupos
    order = ["CRIANÇA (0–11)", "ADOLESCENTE (12–17)", "JOVEM (18–29)", "ADULTO (30–59)", "IDOSO (60+)", "SEM INFO"]
    tab = tab.reindex(index=order, columns=order).fillna(0)

    if normalize == "row":
        tab_plot = tab.div(tab.sum(axis=1).replace(0, np.nan), axis=0) * 100
        title_norm = "% dentro da faixa do suspeito"
        fmt = ".1f"
        annot = True
    elif normalize == "all":
        tab_plot = tab / tab.values.sum() * 100
        title_norm = "% do total"
        fmt = ".2f"
        annot = True
    else:
        tab_plot = tab
        title_norm = "contagem"
        fmt = "d"
        annot = True

    plt.figure(figsize=(9, 6))
    ax = sns.heatmap(tab_plot, cmap="magma", annot=annot, fmt=fmt, linewidths=0.3)
    plt.title(f"Grupo etário do suspeito × Grupo etário da vítima")
    plt.xlabel("Grupo etário da vítima")
    plt.ylabel("Grupo etário do suspeito")
    plt.xticks(rotation=30, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
def cols_like(df, prefix, contains_any=None, exclude_any=None):
    cols = [c for c in df.columns if c.startswith(prefix)]
    if contains_any:
        cols = [c for c in cols if any(x in c for x in contains_any)]
    if exclude_any:
        cols = [c for c in cols if not any(x in c for x in exclude_any)]
    return cols

def flag_from_any(df, cols):
    if not cols:
        return 0
    return df[cols].fillna(0).astype(int).any(axis=1).astype(int)

def safe_get01(df, col):
    if col not in df.columns:
        return 0
    return df[col].fillna(0).astype(int).clip(0,1)


out = dfClusteredB.copy()

# -----------------------
# Canal de atendimento
# -----------------------
canal_digital = [
    "APLICATIVO", "E-MAIL", "PORTAL", "WEBCHAT", "WHATSAPP", "TELEGRAM"
]

canal_presencial = [
    "PRESENCIAL", "CARTA", "AUTUAÇÃO DE OFÍCIO", "PROCESSO SEI"
]

canal_telefonico = ["TELEFÔNICO"]

out["canal_digital"] = flag_from_any(
    out, cols_like(out, "canal_de_atendimento_", contains_any=canal_digital)
)

out["canal_presencial"] = flag_from_any(
    out, cols_like(out, "canal_de_atendimento_", contains_any=canal_presencial)
)

out["canal_telefonico"] = flag_from_any(
    out, cols_like(out, "canal_de_atendimento_", contains_any=canal_telefonico)
)

out["canal_nao_informado"] = safe_get01(out, "canal_de_atendimento_nan")


# -----------------------
# Denunciante
# -----------------------
out["denunciante_vitima"] = safe_get01(out, "denunciante_A PRÓPRIA VÍTIMA")
out["denunciante_agressor"] = safe_get01(out, "denunciante_O PRÓPRIO AGRESSOR")
out["denunciante_terceiro"] = safe_get01(out, "denunciante_TERCEIRO")
out["denunciante_nao_info"] = safe_get01(out, "denunciante_nan")



# -----------------------
# Cenário da violação
# -----------------------
domicilio = ["CASA"]
institucional = [
    "INSTITUIÇÃO", "ESCOLA", "CRECHE", "ILPI",
    "DELEGACIA", "CENTRO DE REFERÊNCIA", "UNIDADE PRISIONAL"
]
via_publica = ["VIA PÚBLICA", "ÔNIBUS", "METRÔ", "TREM"]
virtual = ["AMBIENTE VIRTUAL", "INTERNET"]

out["cenario_domicilio"] = flag_from_any(
    out, cols_like(out, "cenario_da_violacao_", contains_any=domicilio)
)

out["cenario_institucional"] = flag_from_any(
    out, cols_like(out, "cenario_da_violacao_", contains_any=institucional)
)

out["cenario_via_publica"] = flag_from_any(
    out, cols_like(out, "cenario_da_violacao_", contains_any=via_publica)
)

out["cenario_virtual"] = flag_from_any(
    out, cols_like(out, "cenario_da_violacao_", contains_any=virtual)
)


# -----------------------
# Frequência
# -----------------------
out["freq_unica"] = safe_get01(out, "frequencia_ÚNICA OCORRÊNCIA")
out["freq_ocasional"] = safe_get01(out, "frequencia_OCASIONALMENTE")
out["freq_semanal"] = safe_get01(out, "frequencia_SEMANALMENTE")
out["freq_diaria"] = safe_get01(out, "frequencia_DIARIAMENTE")
out["freq_nao_info"] = safe_get01(out, "frequencia_NÃO SABE INFORMAR")


# -----------------------
# Grupo vulnerável
# -----------------------
out["gv_mulher"] = safe_get01(out, "grupo_vulneravel_VIOLÊNCIA CONTRA A MULHER")
out["gv_crianca"] = safe_get01(out, "grupo_vulneravel_VIOLÊNCIA CONTRA CRIANÇA OU ADOLESCENTE")
out["gv_idoso"] = safe_get01(out, "grupo_vulneravel_VIOLÊNCIA CONTRA PESSOA IDOSA")
out["gv_pcd"] = safe_get01(out, "grupo_vulneravel_VIOLÊNCIA CONTRA PESSOA COM DEFICIÊNCIA")
out["gv_lgbtq"] = safe_get01(out, "grupo_vulneravel_VIOLÊNCIA CONTRA POPULAÇÃO LGBTQIA+")


# -----------------------
# Tipo de violação (macro)
# -----------------------
out["viol_fisica"] = flag_from_any(
    out, cols_like(out, "violacao_n3_", contains_any=["FÍSICA", "LESÃO"])
)

out["viol_psiquica"] = flag_from_any(
    out, cols_like(out, "violacao_n3_", contains_any=["PSÍQUICA", "AMEAÇA", "ASSÉDIO"])
)

out["viol_sexual"] = flag_from_any(
    out, cols_like(out, "violacao_n3_", contains_any=["SEXUAL", "ESTUPRO"])
)

out["viol_institucional"] = flag_from_any(
    out, cols_like(out, "violacao_n1_", contains_any=["INSTITUCIONAL"])
)

# -----------------------
# UF (por região)
# -----------------------
uf_norte = ["AC", "AP", "AM", "PA", "RO", "RR", "TO"]
uf_nordeste = ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"]
uf_centro_oeste = ["DF", "GO", "MT", "MS"]
uf_sudeste = ["ES", "MG", "RJ", "SP"]
uf_sul = ["PR", "RS", "SC"]

out["uf_norte"] = flag_from_any(out, cols_like(out, "uf_", contains_any=uf_norte))
out["uf_nordeste"] = flag_from_any(out, cols_like(out, "uf_", contains_any=uf_nordeste))
out["uf_centro_oeste"] = flag_from_any(out, cols_like(out, "uf_", contains_any=uf_centro_oeste))
out["uf_sudeste"] = flag_from_any(out, cols_like(out, "uf_", contains_any=uf_sudeste))
out["uf_sul"] = flag_from_any(out, cols_like(out, "uf_", contains_any=uf_sul))

# Casos especiais (se existirem no seu dataset)
out["uf_atendimento_interrompido"] = safe_get01(out, "uf_ATENDIMENTO INTERROMPIDO")
out["uf_denunciante_nao_soube"] = safe_get01(out, "uf_DENUNCIANTE NÃO SOUBE INFORMAR")
out["uf_nan"] = safe_get01(out, "uf_nan")

# (Opcional) Um flag geral "UF desconhecida"
out["uf_desconhecida"] = (
    out["uf_atendimento_interrompido"] |
    out["uf_denunciante_nao_soube"] |
    out["uf_nan"]
).astype(int)
```
